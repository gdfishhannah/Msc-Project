```python
import numpy as np
import os, sys, time
import collections
import matplotlib.pyplot as plt
```

```python
top_artist = np.load("./eval/top100artist.npy")
genres = np.load('eval/top100artistgenres.npy').item()
```

```python
score_list = []

for num_pred in range(1,100):
    count = 0
    for ip in top_artist:
        related = top_artist[:num_pred]
        ip_gen = genres[ip]
        for j in related:
            for i in genres[j]:
                if i in ip_gen:
                    count+=1
                    break
    score_list.append(count/(100*num_pred))
plt.plot(score_list)
plt.show()
```

```python
score_list = []

for num_pred in range(1,100):
    count = 0
    for ip in top_artist:
        related = top_artist[:num_pred]
        ip_gen = genres[ip]
        correct = 0
        for j in related:
            for i in genres[j]:
                if i in ip_gen:
                    correct = 1
        count += correct
    score_list.append(count/(100))
plt.xticks(np.arange(0, 101, 10))
plt.plot(score_list)
plt.show()
```

```python
np.save('eval/base_acc.npy',score_list)
```
