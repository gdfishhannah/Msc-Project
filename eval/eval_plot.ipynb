```python
import numpy as np
import matplotlib.pyplot as plt
# import seaborn
```

```python
count = np.load('count_eval.npy')
i2v = np.load('i2v_eval.npy')
mc = np.load('mc_eval.npy')
item = np.load('item_eval.npy')
base = np.load('base_eval.npy')
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=128, facecolor='w', edgecolor='k')
plt.plot(base)
plt.plot(count)
plt.plot(i2v)
plt.plot(item)
plt.plot(mc)
plt.legend(["baseline","count based","item2vec","item-based CF","MC"])
plt.xlabel("Number of Prediciton")
plt.ylabel("Accuracy")
plt.title("Algorithm Evaluation")
plt.xticks(np.arange(0, 101, 10))
# plt.grid(True)
plt.show()
```
